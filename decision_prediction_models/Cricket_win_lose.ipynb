```
# Importing the required Libraries
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from google.colab import files
```

```
# Uploading the dataset
files.upload()
```

```
# Reading the Dataframe
dataframe = pd.read_csv('odi_win_lose.csv')
```

```
# Displaying contents of the dataframe
dataframe.head()
```

```
# generating the target labels as 0, 1
# Storing the contents of the main dataframe into an array
dataframe_values = dataframe.values

# Creating an empty target variable
y = []
for i in range(dataframe_values.shape[0]):
  
  # If Winner is equals to Team1
  if (dataframe_values[i][7] == dataframe_values[i][2]):
    y.append(0)
  # Else If Winner is equals to Team2
  elif (dataframe_values[i][7] == dataframe_values[i][3]):
    y.append(1)
```

```
# Storing contents of x in seperate dataframe
x_dataframe = pd.DataFrame(dataframe, columns=['Team 1', 'Team 2', 'Ground', 'Host_Country', 'Venue_Team1', 'Venue_Team2', 'Innings_Team1', 'Innings_Team2'])
```

```
y
```

```
# Label Encoding all the strings in the input
# Define all the label Encoders
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
le4 = LabelEncoder()

# Label Encoding all the features
x_dataframe['Team 1'] = le1.fit_transform(x_dataframe['Team 1'])
x_dataframe['Team 2'] = le1.fit_transform(x_dataframe['Team 2'])
x_dataframe['Host_Country'] = le1.fit_transform(x_dataframe['Host_Country'].astype(str))
x_dataframe['Ground'] = le2.fit_transform(x_dataframe['Ground'])
x_dataframe['Venue_Team1'] = le3.fit_transform(x_dataframe['Venue_Team1'])
x_dataframe['Venue_Team2'] = le3.fit_transform(x_dataframe['Venue_Team2'])
x_dataframe['Innings_Team1'] = le4.fit_transform(x_dataframe['Innings_Team1'])
x_dataframe['Innings_Team2'] = le4.fit_transform(x_dataframe['Innings_Team2'])
```

```
# Finally transferring those values into an array
x = x_dataframe.values

# Splitting the data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```
y_train
```

```
# Building the model
model = Sequential()
model.add(Dense(input_dim=(x_train.shape[1]), units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compiling loss
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
model.summary()
```

```
# Train the model
model.fit(x_train, y_train, epochs=2000)
```

```
# Saving the model
model.save('model.h5')
```

```
# Upload the saved model *Optional 
files.upload()
```

```
# Load model
saved_model = load_model('cricket_win_lose_weights.h5')
```

```
# Predict on test data for evaluation
predictions = saved_model.predict(x_test)
```

```
# Convert the probabilities into discreate labels 0 or 1
predict = []
for i in range(len(predictions)):
  if predictions[i]>=0.5:
    predict.append(1)
  else:
    predict.append(0)
```

```
# Calculate accuracy
a=0
for i in range (len(predict)):
  if (predict[i]==y_test[i]):
    a+=1
validation_score = (a/len(predict))*100
```

```
print(validation_score)
```
